List Operations

```python
# Create a list of 5 integers
numbers = [40, 10, 30, 20, 50]
print("Initial list:", numbers)

# append: adds 60 to the end of the list
numbers.append(60)
print("After append(60):", numbers)

# extend: adds [70, 80] to the end of the list
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

# insert: inserts 25 at index 2
numbers.insert(2, 25)
print("After insert(2, 25):", numbers)

# remove: removes the first occurrence of 30
numbers.remove(30)
print("After remove(30):", numbers)

# pop: removes and returns the last element
popped_value = numbers.pop()
print("After pop():", numbers, "(Popped value was:", popped_value, ")")

# index: finds the index position of the value 20
position = numbers.index(20)
print("Index of 20:", position)

# count: counts how many times 50 appears in the list
occurrences = numbers.count(50)
print("Count of 50:", occurrences)

# sort: sorts the list in ascending order
numbers.sort()
print("After sort():", numbers)

# reverse: reverses the list order
numbers.reverse()
print("After reverse():", numbers)

# clear: removes all elements from the list
numbers.clear()
print("After clear():", numbers)
```

Tuple Mutability

```python
# Create a tuple of 3 student names
students = ("Amit", "Pooja", "Raj")
print("Initial tuple:", students)

# Try changing the second name ('Pooja' to 'Neha')
try:
    students[1] = "Neha"
except TypeError as error:
    print("\n--- Error Occurred ---")
    print(error)
    print("\n--- Explanation ---")
    print("Tuples in Python are IMMUTABLE. Once a tuple is created,")
    print("its elements cannot be changed, added, or removed.")
```

Nested Conditions

```python
# Set number variable
number = 12

if number > 0:
    # Nested check for positive number
    if number % 2 == 0:
        print(number, "is positive and even.")
    else:
        print(number, "is positive but odd.")
elif number == 0:
    print("The number is zero.")
else:
    print(number, "is negative.")
```

Section C: Exploring Pandas

DataFrame Operations

```python
import numpy as np
import pandas as pd

# 1. Create a DataFrame manually with 10 rows (includes some missing values and a duplicate row to demonstrate cleanup)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack', 'Alice'],
    'Age': [25, 30, 35, 40, 22, 28, 33, 45, 29, 31, 25],
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Pune', 'Kolkata', 'Hyderabad', 'Ahmedabad', 'Jaipur', 'Lucknow', 'Mumbai'],
    'Salary': [50000.0, 60000.0, np.nan, 80000.0, 45000.0, np.nan, 70000.0, 95000.0, 55000.0, 65000.0, 50000.0]
}
df = pd.DataFrame(data)
print("--- Original DataFrame (with missing values and a duplicate row) ---")
print(df)

# 2. Use .info()
print("\n--- .info() Output ---")
df.info()

# 3. Use .describe()
print("\n--- .describe() Output ---")
print(df.describe())

# 4. Select 'Name' and 'City' columns
name_and_city = df[['Name', 'City']]
print("\n--- Selected Name and City Columns ---")
print(name_and_city)

# 5. Drop 'City' column
df_no_city = df.drop(columns=['City'])
print("\n--- DataFrame after dropping City column ---")
print(df_no_city)

# 6. Fill missing values in 'Salary' column with the mean
mean_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(mean_salary)
print("\n--- DataFrame after filling missing Salaries with Mean (₹" + str(round(mean_salary, 2)) + ") ---")
print(df)

# 7. Remove any duplicate rows
df_cleaned = df.drop_duplicates()
print("\n--- Final Cleaned DataFrame (No duplicates) ---")
print(df_cleaned)
```

Exercise 1: Prime Numbers


```python
number = int(input("Enter a number: "))
is_prime = True
if number <= 1:
    is_prime = False
else:
    for i in range(2, number):
        if number % i == 0:
            is_prime = False
            break
if is_prime:
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```

Exercise 2: Product of Random Numbers


```python
import random
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)
correct_answer = num1 * num2
print("What is", num1, "multiplied by", num2, "?")
user_answer = int(input("Your answer: "))
if user_answer == correct_answer:
    print("Correct!")
else:
    print("Incorrect. The correct answer was", correct_answer)
```

Exercise 3: Squares of Even/Odd Numbers


```python

for number in range(100, 201):
    # Check if the number is even
    if number % 2 == 0:
        square = number * number
        print("Square of", number, "is", square)
```

Exercise 4: Hospital Billing System


```python
days_stayed = 4
lab_tests = 3
medicine_charges = 2400
room_charges = days_stayed * 2000
doctor_fee = 1500
lab_charges = lab_tests * 300
total_bill = room_charges + doctor_fee + lab_charges + medicine_charges
if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0.0
final_bill = total_bill - discount
print("--- Hospital Bill ---")
print("Room Charges: ₹", room_charges)
print("Doctor Fee: ₹", doctor_fee)
print("Lab Charges: ₹", lab_charges)
print("Medicine Charges: ₹", medicine_charges)
print("---------------------")
print("Total before discount: ₹", total_bill)
print("Discount (10%): ₹", discount)
print("---------------------")
print("Final Bill Amount: ₹", final_bill)
```

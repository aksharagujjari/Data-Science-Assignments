Import Libraries

```python
import numpy as np
from scipy.stats import norm
```

Given Data

```python
# Given values
sample_mean = 3050
X_mean = 600
sigma_X = 25
n = 25

# Theoretical cost model
mu = 1000 + 5 * X_mean

# Standard deviation of weekly cost
sigma = 5 * sigma_X

print("Theoretical Mean Cost =", mu)
print("Standard Deviation =", sigma)
```

State Hypotheses

```python
print("Null Hypothesis (H0): Weekly operating cost follows the model.")
print("Alternative Hypothesis (H1): Weekly operating cost is higher than the model.")
```

Calculate Test Statistic

```python
z = (sample_mean - mu) / (sigma / np.sqrt(n))

print("Z-Test Statistic =", z)
```

Critical Value and P-Value

```python
alpha = 0.05

critical_value = norm.ppf(1 - alpha)

p_value = 1 - norm.cdf(z)

print("Critical Value =", critical_value)
print("P-value =", p_value)
```

Decision

```python
if z > critical_value:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
